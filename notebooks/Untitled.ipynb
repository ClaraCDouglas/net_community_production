```python
#flip data
pres=data.Pressure[:, ::-1]
temp=data.Temperature[:, ::-1]

# Quick Timeseries Profile plot of Temperature
temp.T.plot(cmap='RdBu_r')
plt.gca().invert_yaxis()
```
